```python
# visualizations
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline 
```

# Read in Dataset

```python
df = pd.read_csv("DataSource/master.csv", index_col=0)
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
sub_df = df[['acousticness', 'danceability', 'duration_ms', 'energy',
       'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo',
       'valence', 'popularity', 'key', 'mode', 'count','clean_genres']]
sub_df.head()
```

```python
feature = "acousticness"
```

```python
sns.set(color_codes=True)
plt.figure(figsize=(15,10))
sns.distplot(sub_df[feature].loc[(sub_df["clean_genres"] == 'show tunes')], hist=False, color="darkred", label="show tunes")
sns.distplot(sub_df[feature].loc[(sub_df["clean_genres"] == 'electronic')], hist=False, color="deeppink", label="electronic")
sns.distplot(sub_df[feature].loc[(sub_df["clean_genres"] == 'worship')], hist=False, color="lightcoral", label="worship")
sns.distplot(sub_df[feature].loc[(sub_df["clean_genres"] == 'pop')], hist=False, color="darkorange", label="pop")
sns.distplot(sub_df[feature].loc[(sub_df["clean_genres"] == 'jazz')], hist=False, color="gold", label="jazz")
sns.distplot(sub_df[feature].loc[(sub_df["clean_genres"] == 'rock')], hist=False, color="darkgreen", label="rock")
sns.distplot(sub_df[feature].loc[(sub_df["clean_genres"] == 'rap')], hist=False, color="yellowgreen", label="rap")
sns.distplot(sub_df[feature].loc[(sub_df["clean_genres"] == 'classical')], hist=False, color="lightseagreen", label="classical")
sns.distplot(sub_df[feature].loc[(sub_df["clean_genres"] == 'country')], hist=False, color="lime", label="country")
sns.distplot(sub_df[feature].loc[(sub_df["clean_genres"] == 'latin')], hist=False, color="dodgerblue", label="latin")
sns.distplot(sub_df[feature].loc[(sub_df["clean_genres"] == 'r&b')], hist=False, color="mediumslateblue", label="r&b")
sns.distplot(sub_df[feature].loc[(sub_df["clean_genres"] == 'indie')], hist=False, color="indigo", label="indie")
plt.title(f"Spotify Audio Feature Distibution by Genre ({feature})", fontsize= 15)
plt.savefig(f'Images/Histogram_Feature_Distribution_by_Genre_{feature}.jpg', format="jpeg", dpi=300)
plt.show()
```

```python
def plotFeatureDistGenre(df,feature):
    
    plt.figure(figsize=(15,10))
    sns.distplot(sub_df[feature].loc[(df["clean_genres"] == 'show tunes')], hist=False, color="darkred", label="show tunes")
    sns.distplot(sub_df[feature].loc[(df["clean_genres"] == 'electronic')], hist=False, color="deeppink", label="electronic")
    sns.distplot(sub_df[feature].loc[(df["clean_genres"] == 'worship')], hist=False, color="lightcoral", label="worship")
    sns.distplot(sub_df[feature].loc[(df["clean_genres"] == 'pop')], hist=False, color="darkorange", label="pop")
    sns.distplot(sub_df[feature].loc[(df["clean_genres"] == 'jazz')], hist=False, color="gold", label="jazz")
    sns.distplot(sub_df[feature].loc[(df["clean_genres"] == 'rock')], hist=False, color="darkgreen", label="rock")
    sns.distplot(sub_df[feature].loc[(df["clean_genres"] == 'rap')], hist=False, color="yellowgreen", label="rap")
    sns.distplot(sub_df[feature].loc[(df["clean_genres"] == 'classical')], hist=False, color="lightseagreen", label="classical")
    sns.distplot(sub_df[feature].loc[(df["clean_genres"] == 'country')], hist=False, color="lime", label="country")
    sns.distplot(sub_df[feature].loc[(df["clean_genres"] == 'latin')], hist=False, color="dodgerblue", label="latin")
    sns.distplot(sub_df[feature].loc[(df["clean_genres"] == 'r&b')], hist=False, color="mediumslateblue", label="r&b")
    sns.distplot(sub_df[feature].loc[(df["clean_genres"] == 'indie')], hist=False, color="indigo", label="indie")
    plt.title(f"Spotify Audio Feature Distibution by Genre ({feature})", fontsize= 15)
    plt.savefig(f'Images/Histogram_Feature_Distribution_by_Genre_{feature}.jpg', format="jpeg", dpi=300)
    plt.show()
```
