# Import Dependencies

```python
# visualizations
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline 
```

# Read in Dataset

```python
df = pd.read_csv("DataSource/master.csv", index_col=0)
df.head()
```

```python
genre_count = df.groupby(by="clean_genres").count()["artists"].to_frame()
genre_count.rename(columns={"artists": "Number of Tracks"}, inplace=True)
genre_count.rename_axis("Genre", inplace=True)
genre_count.sort_values(by="Number of Tracks", ascending=False, inplace=True)
genre_count.reset_index(inplace=True)
# type(genre_count)
genre_count
```

```python
sns.set()
sns.catplot(y="Genre", x="Number of Tracks", palette="rocket", 
            data=genre_count, kind="bar", height=6, aspect=3)
# plt.savefig(f'Images/Top 10 Tracks by Streams (Spotify Top 200 Tracks {region}).jpg', format="jpeg", dpi=300)
plt.title(f"Number of Tracks by Genre", fontsize= 15)
plt.show()
```
